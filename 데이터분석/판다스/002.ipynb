```python
import pandas as pd
data = {
    'name': ['Kim', 'Lee', 'Park'],
    'age': [25, 30, 35],
    'city': ['Seoul', 'Busan', 'Incheon']
}
df = pd.DataFrame(data)
df
```

```python
data = [
    ['Kim', 25, 'Seoul'],
    ['Lee', 30, 'Busan'],
    ['Park', 35, 'Incheon']
]
df = pd.DataFrame(data, 
                   columns=['name', 'age', 'city']
                  )
df
```

```python
s1 = pd.Series(['Kim', 'Lee', 'Park'])
s2 = pd.Series([25, 30, 35],index=['a','b','c'])
df = pd.DataFrame({'name': s1, 'age': s2})
df
```

```python
# df.shape        # (행 수, 열 수)
# df.size         # 전체 요소 개수
# df.ndim         # 차원 (항상 2)
# df.index        # 인덱스 객체
# df.columns      # 컬럼 객체
# df.values       # NumPy 배열
# df.dtypes       # 각 컬럼의 데이터 타입
data = {
    'city' : ['서울','대전','대구','부산'],
    'data1' : np.random.randint(100000,1000000,4),
    'data2' : np.random.randn(4),
}
df = pd.DataFrame(data)
df
print(df.shape)  # ( 4 , 3 )
print(df.size) # 12
print(df.ndim)
print(list(df.index))
print(df.columns)
print(df.values)
print(df.dtypes)
```
