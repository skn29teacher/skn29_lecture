```python
import pandas as pd
s1 = pd.Series([1, 2, 3,4], index=['a', 'b', 'c','d'])
s2 = pd.Series([10, 20, 30,40], index=['a', 'bb', 'c','dd'])
s = s1+s2
s.fillna(s.mean())
```

```python
s.sort_values(ascending=False)
s.sort_index()
```

```python
# 예제 1: 학생 성적 관리
# 5명의 학생 성적을 Series로 생성
# 평균, 최고점, 최저점 계산
# 80점 이상 학생 필터링
```

```python
data = {
    '홍길동' : 95,
    '이순신' : 75,
    '강감찬' : 94,
    '철수' : 77,
    '철이' : 90,
}
s = pd.Series(data)

pd.Series( [s.mean(),s.max(),s.min()], index=['평균','최대값','최소값'])

s[s>=80]
```

```python
# 주요 도시의 인구를 딕셔너리로 Series 생성
population = pd.Series({
    '인천': 2947217,
    '대구': 2427954,
    '서울': 9720846,
    '부산': 3404423,    
    '대전': 1471040
})
population

# 인구 순으로 정렬
population.sort_values(ascending=False)
# 특정 인구 범위 필터링
population[population >=2500000]
```
