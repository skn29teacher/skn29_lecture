```python
import pandas as pd
from glob import glob

filepath = 'data/*.csv'
files = glob(filepath)
print(files[8])
df = pd.read_csv(files[8])
df.head()
```

```python
df2 = df.drop('상가업소번호',axis=1)
df2.columns
```

```python
result = []
for cname in df2.columns:
    if '코드' in cname:
        result.append(cname)

result        
```
