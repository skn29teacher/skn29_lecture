```python
# 퍼셉트론
# 입력(x1 x2 x3 ..)
# 가중치(w1 w2 w3 ..)
# 가중합(w1*x1 + w2*x2 + w3*x3 + ...)
# 활성화 함수(activation function) : 가중합이 임계값(threshold) 이상이면 1, 그렇지 않으면 0 활성화함수는
# 계단함수(step function)로 표현할 수 있음
# 퍼셉트론 학습 알고리즘 : 가중치와 임계값을 조정하여 퍼셉트론이 올바른 출력을 내도록 학습하는 알고리즘
# 퍼셉트론은 선형 분류기(linear classifier)로, 선형적으로 구분 가능한 문제에 대해서만 올바른 출력을 낼 수 있음
# XOR 문제는 선형적으로 구분 불가능한 문제로, 퍼셉트론으로는 해결할 수 없음
```

```python
# 데이터 만들기
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
X,y = make_blobs(
    n_samples=100,
    centers=2,
    random_state=42,
    cluster_std=1.8
)
fig, ax = plt.subplots(figsize=(6,5))
ax.scatter(X[:,0], X[:,1], c=y, cmap='bwr')
plt.show()
```

```python
# w1*x1 + w2*x2 + b = 0   직선의 공식
import numpy as np
np.random.seed(100)
w = np.random.randn(2)
b = np.random.randn()
w,b
```

```python
# y = f(x) = ax+b
# 수식을 변형
# w1*x1 + w2*x2 + b = 0   y = ?  
# W1*x + W2*y + b = 0
# W2*y = -(W1*x + b)
# y = -(W1*x + b) / W2
```

```python
# 직선을 그리는 함수
def plot_line(w,b):
    x_vals = np.linspace(X[:,0].min(), X[:,1].max(),100)
    y_vals = -(w[0]*x_vals + b) / w[1]
    _,ax = plt.subplots(figsize = (6,5))
    ax.scatter(X[:,0], X[:,1],c=y, cmap='bwr')
    ax.plot(x_vals, y_vals,color='black')
    ax.set_title('Current Decision Boundary')
    plt.show()

plot_line(w,b)
```

```python
# 틀린점 하나 발견
def preceptron_step(X,y,w,b,lr=0.1):
    for i in range(len(X)):
        xi = X[i]
        yi = y[i]
        yi = 1 if yi == 1 else -1
        pred = np.sign(np.dot(w,xi) +b )
        if pred != yi:
            w = w + lr * yi * xi
            b = b + lr*yi
            return w, b
    return w,b 
# 한번 업데이트
w,b = preceptron_step(X,y,w,b)
plot_line(w,b)
```

```python
# 반복
for i in range(20):
    w,b = preceptron_step(X,y,w,b)    
```

```python
plot_line(w,b)
```
