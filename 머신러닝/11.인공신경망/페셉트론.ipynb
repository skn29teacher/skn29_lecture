## 학습  - 오차를 줄이는 방향

wi​←wi​+η⋅(y−y^​)⋅xi​

```python
# 정답 1 예측 0 (과소)  오차  +1
# w_new = w_old + lr*(y-y_hat)*x
0.3+0.5*(1-0)*1  # 0.8
# w_old 0.3
# w_new 0.8
# delta_w  0.5

# 정답0 예측 1 (과대)  오차 -1
0.3+0.5*(0-1)*1  # -0.2
# w_old 0.3
# w의 변화량 -0.5

# 정답 1 예측 1
0.3+0.5*(1-1)*1  # 0.3
# w_old 0.3
# w의 변화량 0
```

```python
x = np.array([[0,0],[1,1],[0,1],[1,0]])
y = np.array([0,1,1,0])

# 가중치 편향을 
import numpy as np
np.random.seed(42)
w = np.random.randn(2)
b = np.random.rand()
lr = 0.1

# 학습루프 100
for epoch in range(100):
    for xi, yi in zip(x,y):
        z = np.dot(w,xi) + b
        y_hat = 1 if z >=0 else 0
        error = yi - y_hat
        w +=  lr*error*xi 
        b += lr*error

print(f'학습후 w : {w} b : {b}')

new_x = np.array([[0,0],[1,1],[0,1],[1,0]])
for n_x,y_ in zip(new_x,y):
    z = np.dot(w,n_x) + b
    y_hat = 1 if z >=0 else 0
    print(f'{n_x}에한 y의 값 예측 {y_hat} 정답 : {y_}')
```
