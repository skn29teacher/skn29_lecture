평가지표 MAE MSE RMSE $R^2$

$$ MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i| $$

- 모든 오차에 동일한 가중치를 적용
- 이상치에 덜 민감한 지표가 필요할때 , 오차를 직관적으로 파악하기 쉽다

$$ MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $$

- 오차를 제곱 : 1미만의 오차는 작아지고 1이상의 오차는 매우커짐
- 모델이 오차가 큰 예측에 대해 큰 패널티를 부여 - 주로 사용
- 이상치를 제거하는게아니라 이상치가 중요한 도메인 : 금융리스크, 의료진단

```python
0.5**2, 1.5**2   ,1.5-0.5,  2.25-0.25
```

$$ RMSE = \sqrt{MSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2} $$

- MSE에 제곱을 했으니 원래 데이터의 SCALE이 변함 그래서 맞춤
- MSE의 장점을 그대로 가져가면서 MAE 장점도 가져감(오차가큰 가중치에 패널티를 더 크게하면서 오차를 직관적으로 파악)

$$ R^2 = 1 - \frac{SSE}{SST} = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2} $$

- 데이터의 분산을  설명하는 상대적인 성능(0 ~ 1)
- SSE(Sum of Squred Errors) : 오차의 제곱의 합  (잔차의 합)
- SST(Total Sum of Squares) : 
    - 실제값들의 평균값( $\bar{y}$ )과 실제값들의 차이 제곱합 - 편차들의 제곱합
    - 실제값들의 평균값( $\bar{y}$ )과 실제값들의 차이 제곱합 이후 평균 - 분산

- $\bar{y}$ : (10+20+30) / 3 = 20
- 실제 : 10 20 30 
- 예측 : 12 18 33
- SST :  $(10-20)^2$ = 100
    - 100 + 0 + 100 = 200
- SSE : 17
- 1- 14/200 = 0.915

- 0~1사이의 측정지표  객관적인 성능비교

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
x_train,x_test,y_train,y_test = train_test_split(X,y,random_state=42)
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from xgboost import XGBRegressor
xgb = XGBRegressor()
xgb.fit(x_train,y_train)
r2_score(y_test, xgb.predict(x_test))
```

```python
import torch
import torch.nn as nn
from torch.optim import Adam
from torch.utils.data import TensorDataset, DataLoader
x_train_t = torch.FloatTensor(x_train)
x_test_t = torch.FloatTensor(x_test)
y_train_t = torch.FloatTensor(y_train)
y_test_t = torch.FloatTensor(y_test)
train_dataset = TensorDataset(x_train_t,y_train_t)
test_dataset = TensorDataset(x_test_t,y_test_t)
train_loader = DataLoader(train_dataset,batch_size=36,shuffle=True)
test_loader = DataLoader(test_dataset,batch_size=32)
```

```python
class IrisReg(nn.Module):
    def __init__(self, input_dim, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.network = nn.Sequential(            
            nn.Linear(input_dim, 64),
            nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(64, 32),
            nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(32, 1)            
        )
    def forward(self,x):
        return self.network(x)

```
